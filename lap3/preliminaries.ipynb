# Kernel methods

Consider linear regression in feature space with regularization (kernel ridge regression).

$$
J(w)=\frac{1}{2}\sum_{n=1}^{N}\big(w^T\phi(x_n)-t_n\big)^2+\frac{\lambda}{2}\lVert w\rVert^2,
$$

with $\lambda>0$. In matrix form:

$$
J(w)=\frac{1}{2}\lVert \Phi w-t\rVert^2+\frac{\lambda}{2}\lVert w\rVert^2,
$$

where $\Phi\in\mathbb{R}^{N\times M}$ (row $n$ is $\phi(x_n)^T$) and $t=(t_1,\dots,t_N)^T$.

Setting the gradient to zero:

$$
\nabla_w J(w)=\Phi^T(\Phi w-t)+\lambda w=0
\quad\Longrightarrow\quad
(\Phi^T\Phi+\lambda I_M)w=\Phi^T t.
$$

By the representer form, write $w=\Phi^T a$ with $a\in\mathbb{R}^N$. Define Gram matrix:

$$
K=\Phi\Phi^T,\qquad K_{mn}=\phi(x_m)^T\phi(x_n)=k(x_m,x_n).
$$

Substitute $w=\Phi^Ta$ into the objective:

$$
J(a)=\frac{1}{2}\lVert Ka-t\rVert^2+\frac{\lambda}{2}a^TKa.
$$

Its gradient is:

$$
\nabla_a J(a)=K(Ka-t)+\lambda Ka=K\big((K+\lambda I_N)a-t\big).
$$

Hence (for $\lambda>0$) the dual coefficients satisfy:

$$
(K+\lambda I_N)a=t
\quad\Longrightarrow\quad
a=(K+\lambda I_N)^{-1}t.
$$

Prediction for a new input $x$:

$$
y(x)=w^T\phi(x)=a^T k(x)=k(x)^T(K+\lambda I_N)^{-1}t,
$$

where $k(x)=(k(x_1,x),\dots,k(x_N,x))^T$.

This is the **dual representation** of regularized linear regression in kernel form.

# Support Vector Machine

## Formation

For binary classification with labels $a_n\in\{-1,+1\}$, use the linear decision function

$$
f(x)=w^Tx+b.
$$

The decision boundary is the hyperplane

$$
\mathcal{H}=\{x\mid w^Tx+b=0\},
$$

and $w$ is normal to $\mathcal{H}$.

The signed distance from any point $x$ to the boundary is

$$
\frac{w^Tx+b}{\lVert w\rVert},
$$

so for sample $(x_n,a_n)$ the geometric margin is

$$
\gamma_n=\frac{a_n(w^Tx_n+b)}{\lVert w\rVert}.
$$

For linearly separable data, we scale $(w,b)$ so that the closest points satisfy

$$
a_n(w^Tx_n+b)=1.
$$

Then constraints become

$$
a_n(w^Tx_n+b)\ge 1,\quad n=1,\dots,N,
$$

and maximizing margin is equivalent to

$$
\min_{w,b}\;\frac{1}{2}\lVert w\rVert^2
\quad\text{s.t.}\quad
a_n(w^Tx_n+b)\ge 1\;\forall n.
$$

This is the primal hard-margin SVM optimization problem.

## Dual Representation

Introduce Lagrange multipliers $\lambda_n\ge 0$ for constraints $a_n(w^T\phi(x_n)+b)-1\ge 0$.

The Lagrangian is

$$
L(w,b,\lambda)=\frac{1}{2}\lVert w\rVert^2-\sum_{n=1}^{N}\lambda_n\big(a_n(w^T\phi(x_n)+b)-1\big).
$$

Stationarity conditions:

$$
\frac{\partial L}{\partial w}=0
\Longrightarrow
w=\sum_{n=1}^{N}\lambda_n a_n\phi(x_n),
$$

$$
\frac{\partial L}{\partial b}=0
\Longrightarrow
\sum_{n=1}^{N}\lambda_n a_n=0.
$$

Substituting back gives the dual objective:

$$
\max_{\lambda}\;\tilde L(\lambda)
=\sum_{n=1}^{N}\lambda_n
-\frac{1}{2}\sum_{n=1}^{N}\sum_{m=1}^{N}
\lambda_n\lambda_m a_n a_m k(x_n,x_m),
$$

subject to

$$
\lambda_n\ge 0,\qquad
\sum_{n=1}^{N}\lambda_n a_n=0.
$$

After solving for $\lambda^*$, classification is done by

$$
y(x)=\sum_{n=1}^{N}\lambda_n^* a_n k(x_n,x)+b^*,
$$

and prediction is $\operatorname{sign}(y(x))$.

KKT complementary slackness:

$$
\lambda_n\big(a_n y(x_n)-1\big)=0.
$$

Samples with $\lambda_n>0$ are the **support vectors**.

## Overlapping class distribution

Hard-margin SVM fails when data are not perfectly separable. Soft-margin SVM introduces slack variables $\xi_n\ge 0$:

$$
a_n(w^T\phi(x_n)+b)\ge 1-\xi_n,
\qquad \xi_n\ge 0.
$$

Interpretation:

- $\xi_n=0$: correctly classified and outside/on margin.
- $0<\xi_n<1$: correctly classified but inside margin.
- $\xi_n\ge 1$: misclassified.

Primal soft-margin objective:

$$
\min_{w,b,\xi}\;\frac{1}{2}\lVert w\rVert^2 + C\sum_{n=1}^N \xi_n,
$$

subject to the constraints above.

Lagrangian with multipliers $\lambda_n\ge 0$ and $\mu_n\ge 0$:

$$
L=\frac{1}{2}\lVert w\rVert^2 + C\sum_{n=1}^N\xi_n
-\sum_{n=1}^N\lambda_n\big(a_n(w^T\phi(x_n)+b)-1+\xi_n\big)
-\sum_{n=1}^N\mu_n\xi_n.
$$

Stationarity:

$$
\frac{\partial L}{\partial w}=0
\Longrightarrow
w=\sum_{n=1}^N\lambda_n a_n\phi(x_n),
$$

$$
\frac{\partial L}{\partial b}=0
\Longrightarrow
\sum_{n=1}^N\lambda_n a_n=0,
$$

$$
\frac{\partial L}{\partial \xi_n}=0
\Longrightarrow
C-\lambda_n-\mu_n=0
\Longrightarrow
0\le \lambda_n\le C.
$$

Dual soft-margin problem:

$$
\max_{\lambda}\;\sum_{n=1}^{N}\lambda_n
-\frac{1}{2}\sum_{n=1}^{N}\sum_{m=1}^{N}\lambda_n\lambda_m a_n a_m k(x_n,x_m),
$$

subject to

$$
0\le \lambda_n\le C,
\qquad
\sum_{n=1}^{N}\lambda_n a_n=0.
$$

The decision function remains

$$
y(x)=\sum_{n=1}^{N}\lambda_n a_n k(x_n,x)+b.
$$

## Hinge loss for SVM

Soft-margin SVM can be written as regularized empirical risk with hinge loss.

For each sample:

$$
\ell_n = \max\big(0,\,1-a_n f(x_n)\big),
\qquad f(x)=w^Tx+b.
$$

Overall objective:

$$
\mathcal{L}(w,b)=\frac{1}{2}\lVert w\rVert^2 + C\sum_{n=1}^{N}\max\big(0,\,1-a_n(w^Tx_n+b)\big).
$$

Subgradient with respect to $w$:

$$
\partial_w \mathcal{L}
= w - C\sum_{n:\,a_n f(x_n)<1} a_n x_n.
$$

So optimization can be done by (sub)gradient-based methods (for example SGD or mini-batch subgradient descent).
