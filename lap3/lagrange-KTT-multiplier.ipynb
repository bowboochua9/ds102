## Lagrange Multipliers

**Lagrange multipliers** are used to optimize a function under equality constraints.

Consider:

$$
\max_x f(x)\quad\text{s.t.}\quad g(x)=0,
$$

with $x\in\mathbb{R}^D$.

At a constrained optimum $x^*$ (under regularity assumptions), the gradient of $f$ is parallel to the gradient of the active constraint:

$$
\nabla f(x^*) = -\lambda^*\nabla g(x^*),
$$

for some scalar $\lambda^*\in\mathbb{R}$.

Equivalent form:

$$
\nabla f(x^*) + \lambda^*\nabla g(x^*)=0.
$$

Define the Lagrangian

$$
L(x,\lambda)=f(x)+\lambda g(x).
$$

The stationary conditions are

$$
\nabla_x L(x,\lambda)=\nabla f(x)+\lambda\nabla g(x)=0,
$$

$$
\frac{\partial L(x,\lambda)}{\partial \lambda}=g(x)=0.
$$

So solving $(x^*,\lambda^*)$ from these equations gives candidates for constrained optima.

For multiple equality constraints $g_i(x)=0$, $i=1,\dots,m$:

$$
L(x,\lambda)=f(x)+\sum_{i=1}^{m}\lambda_i g_i(x),
$$

with conditions

$$
\nabla f(x)+\sum_{i=1}^{m}\lambda_i\nabla g_i(x)=0,
\qquad g_i(x)=0\ (i=1,\dots,m).
$$

## Karush - Kuhn - Tucker Multipliers

For inequality constraints, use **Karush-Kuhn-Tucker (KKT)** conditions.

A standard form is:

$$
\min_x f(x)\quad\text{s.t.}\quad g_i(x)\le 0,\ i=1,\dots,m.
$$

Define

$$
L(x,\lambda)=f(x)+\sum_{i=1}^{m}\lambda_i g_i(x),
$$

with multipliers $\lambda_i\ge 0$.

At a local optimum (with suitable constraint qualification), KKT requires:

$$
\nabla_x L(x,\lambda)=\nabla f(x)+\sum_{i=1}^{m}\lambda_i\nabla g_i(x)=0
\quad\text{(stationarity)},
$$

$$
g_i(x)\le 0
\quad\text{(primal feasibility)},
$$

$$
\lambda_i\ge 0
\quad\text{(dual feasibility)},
$$

$$
\lambda_i g_i(x)=0\quad (i=1,\dots,m)
\quad\text{(complementary slackness)}.
$$

Equivalent sign convention: if constraints are written as $h_i(x)\ge 0$, then one can use

$$
L(x,\lambda)=f(x)-\sum_i\lambda_i h_i(x),\quad \lambda_i\ge 0,
$$

and the same logic applies.

Interpretation of complementary slackness:

- If $\lambda_i>0$, then $g_i(x)=0$ (active constraint).
- If $g_i(x)<0$, then $\lambda_i=0$ (inactive constraint).

This framework generalizes Lagrange multipliers from equality-constrained optimization to inequality-constrained optimization.
